### 강성규 PART 

```python
# 연도별 회사 조회 함수

def sorted_year(df, year):
    df = df.query(f'년도 == {year}')
    df = df.sort_values(by=['회사ID', '년도', '월'])
    df = df.reset_index(drop=True)
    return df

df_2015 = sorted_year(wanted_nps_df, 2015)
df_2015
```

```python
# 평균 직원 수 column 추가

# '월', '월별_연금보험료'를 제외한 새로운 DataFrame 생성
employ_avg_df_2015 =  pd.DataFrame(df_2015[['회사ID', '연매출액', '년도']], columns= ['회사ID', '연매출액', '년도', '연평균_직원수'])
employ_avg_df_2015 = employ_avg_df_2015.drop_duplicates(keep = 'first')


# 방법 1: for loop 
avg = []
for id in df_2015['회사ID'].unique():
    result = df_2015.query(f'회사ID == {id}').loc[:, '월별_직원수'].mean()
    avg.append(int(result)) # 사람 수 이기 때문에 소수점 아래는 버림

employ_avg_df_2015['연평균_직원수'] = avg


# 방법 2: list comprehention
# employ_avg_df_2015['연평균_직원수'] = [ int(df_2015.query(f'회사ID == {id}').loc[:, '월별_직원수'].mean()) for id in df_2015['회사ID'].unique() ]


# 연매출액을 연평균_직원 수로 나누고, 내림차순으로 정렬합니다.
employ_avg_df_2015['1인당_매출액(천원)'] = employ_avg_df_2015['연매출액'] // employ_avg_df_2015['연평균_직원수'] # 소수점 버림
employ_avg_df_2015 = employ_avg_df_2015.sort_values(by=['1인당_매출액(천원)'], ascending=False) # ascending=False 내림차순

employ_avg_df_2015.head()
```

```python
# "연평균 직원 수 column 추가" 함수

def employ_avg_extract(df):
    avg_df = pd.DataFrame(df[['회사ID', '연매출액', '년도']], columns= ['회사ID', '연매출액', '년도', '연평균_직원수'])
    avg_df = avg_df.drop_duplicates(keep = 'first')
    
    avg = []
    # 방법 1: for loop 
    for id in avg_df['회사ID'].unique():
        result = df.query(f'회사ID == {id}').loc[:, '월별_직원수'].mean()
        avg.append(int(result)) # 사람 수 이기 때문에 소수점 아래는 버림
    
    avg_df['연평균_직원수'] = avg
    avg_df['1인당_매출액(천원)'] = avg_df['연매출액'] // avg_df['연평균_직원수'] # 소수점 버림
    avg_df = avg_df.sort_values(by=['1인당_매출액(천원)'], ascending=False) # ascending=False 내림차순
    avg_df = avg_df.reset_index(drop=True)
    return avg_df
```

```python
#( 예시 ) 함수를 이용한 데이터 조회 

df_2015 = sorted_year(wanted_nps_df, 2015)
employ_avg_df = employ_avg_extract(df_2015)
employ_avg_df.head()
```

```python
# 연도 별 1인당_매출액 상위 5개 기업을 모은 DataFrame 생성
# 조회하는 데 시간이 다소 소요됩니다.

all_employ_avg_df = pd.DataFrame()
year_list = sorted(wanted_nps_df['년도'].unique())

for year in year_list:
    df_year = sorted_year(wanted_nps_df, year)
    employ_avg_df = employ_avg_extract(df_year)
    all_employ_avg_df = all_employ_avg_df.append(employ_avg_df.head(), ignore_index = True)

all_employ_avg_df.head(10)
```
